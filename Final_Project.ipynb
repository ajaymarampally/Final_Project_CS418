**Import the Libraries Required for the project**

```python
from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json
import csv
```

**Write a function to write the json data to csv**
<br>
<br>

```python
def write_to_csv(json_data,filename):
    '''function to write the json output to csv file'''
    
    #open the file to write csv 
    
    api_data_file = open(filename,'w',newline='')
    csv_writer = csv.writer(api_data_file)
    t_dict = list(json_data.values())
    tt_dict = t_dict[1]
    quote_val = []
    del(tt_dict[23])
    # all dictionary items 
    cnt = 0
    for i in range(1,len(tt_dict)):
        main_dict = tt_dict[i]
        #save the value of quote usd and remove after writing to csv
        t_quote = main_dict['quote']['USD']
        quote_val.append(t_quote)
        del(main_dict['quote']['USD'])
        #remove the quote key from dictionary 
        del main_dict['quote']
        del main_dict['tags']
        if cnt ==0:
            header = main_dict.keys()
            csv_writer.writerow(header)
            cnt+=1
        csv_writer.writerow(main_dict.values())
    cnt2 = 0
    #add the values of the saved quote dictionaries to the csv file
    for z in range(len(quote_val)):
        print('index: ',z)
        main_quote = quote_val[z]
        if cnt2==0:
            header1 = main_quote.keys()
            csv_writer.writerow(header1)
            cnt2+=1
        csv_writer.writerow(main_quote.values())   
    api_data_file.close()
```

**Get the Data from the API (coin-base) and write to CSV file**

```python
url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
parameters = {
  'start':'1',
  'limit':'200',
  'convert':'USD'
}
headers = {
  'Accepts': 'application/json',
  'X-CMC_PRO_API_KEY': '9b695354-825b-4a21-ae0e-7efe986c090b',
}

session = Session()
session.headers.update(headers)

try:
  response = session.get(url, params=parameters)
  data = json.loads(response.text)
  write_to_csv(data,'api_data.csv')
  print(data)
except (ConnectionError, Timeout, TooManyRedirects) as e:
  print(e)
```
